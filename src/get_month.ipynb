```python
import pandas as pd
 
data = pd.read_csv('../temp_data/influA_vietnam_last_10_days.csv')
 
data['Day'] = pd.to_datetime(data['Day'])
 
data['Month'] = data['Day'].dt.to_period('M')
 
monthly_data = data.groupby('Month').agg({
    'Influenza A - All types of surveillance': 'sum',
    'temp': 'mean',
    'humidity': 'mean',
    'dew': 'mean',
    'precipcover': 'mean',
    'tempmax': 'max',
    'tempmin': 'min',
    'windspeed': 'mean'
}).reset_index()
 
monthly_data['Month'] = monthly_data['Month'].dt.to_timestamp()
```
